```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option('display.max_columns', None)
```

```python
df = pd.read_csv("../data/cleaned/cleaned_employee_attrition.csv")

df.head()
```

```python
plt.figure(figsize=(5,5))

sns.countplot(x='Attrition', data=df)

plt.title("Employee Attrition Distribution")
plt.show()
```

```python
plt.figure(figsize=(6,4))

sns.countplot(x='OverTime', hue='Attrition', data=df)

plt.title("Overtime vs Attrition")
plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.boxplot(x='Attrition', y='MonthlyIncome', data=df)

plt.title("Monthly Income vs Attrition")
plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.countplot(x='JobSatisfaction', hue='Attrition', data=df)

plt.title("Job Satisfaction vs Attrition")
plt.show()
```

```python
pd.crosstab(df['OverTime'], df['Attrition'], normalize='index') * 100
```

Employees doing overtime have ~30% attrition,
while employees not doing overtime have only ~10% attrition.

```python
plt.figure(figsize=(8,5))

sns.countplot(x='WorkLifeBalance', hue='Attrition', data=df)

plt.title("Work Life Balance vs Attrition")
plt.show()
```

```python
pd.crosstab(
    df['WorkLifeBalance'],
    df['Attrition'],
    normalize='index'
) * 100
```

Employees with poor work-life balance (1) have 31% attrition, which is almost double compared to balance levels 2–4.

```python
plt.figure(figsize=(8,5))

sns.boxplot(x='Attrition', y='YearsAtCompany', data=df)

plt.title("Years at Company vs Attrition")
plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.boxplot(x='Attrition', y='Age', data=df)

plt.title("Age vs Attrition")
plt.show()
```

```python
pd.crosstab(
    df['Department'],
    df['Attrition'],
    normalize='index'
) * 100
```

```python
pd.crosstab(
    df['JobRole'],
    df['Attrition'],
    normalize='index'
) * 100
```

### Key EDA Insights

1. Employees doing overtime show significantly higher attrition (~30%) compared to employees without overtime (~10%).

2. Employees with poor work-life balance (level 1) exhibit highest resignation tendency (~31%).

3. Lower-income employees tend to leave more frequently.

4. Younger employees and newly joined employees show higher attrition risk.

5. Sales department has higher employee attrition, with Sales Representatives showing the highest quitting tendency.
